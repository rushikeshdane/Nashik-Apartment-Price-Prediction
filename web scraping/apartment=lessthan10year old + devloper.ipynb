```python
import regex as re
def min_price(value):
    
    min_price = re.findall(r"[-+]?\d*\.\d+|\d+",value)
    

    if 'Cr\nEMI starts at'in value:
        value2 = float(min_price[0])*100
        return round(value2,ndigits=1)
    elif 'Price on Request' in value:
        return 'Price on Request'
    else:
        return min_price[0]
```

```python
df['min_price'] = df['emi'].apply(min_price)
```

```python
def BHK1(value):
    
    if 'BHK' in value:
        property_1= re.findall('[0-9]',value)
        return property_1[0]
    else:
        return value
```

```python
df['BHK1'] = df['sq_ft'].apply(BHK1)
```

```python
def max_price(value):
    
    max_price = re.findall(r"[-+]?\d*\.\d+|\d+",value)
    


    if 'Price on Request' in value:
        return 'Price on Request'
    elif len(value) < 30 :
        return "NA"
    elif 'Cr\nEMI starts at'in value:
        value2 = float(max_price[1])*100
        return round(value2,ndigits=1)
    else:
        return max_price[1]
```

```python
df['max_price'] = df['emi'].apply(max_price)
```

```python
def BHK2(value):
    
    if 'BHK'or 'RK'  in value:
        if len(value) < 18:
            return 'NA'
        elif '3 BHK Penthouse, Apartment'==value:
            return 'NA'
        else:
            property_1= re.findall('[0-9]',value)
            return property_1[1]
    else:
        return value
```
